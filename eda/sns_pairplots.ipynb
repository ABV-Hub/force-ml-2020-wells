# Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import re
import hvplot.pandas
import seaborn as sns
from dotenv import load_dotenv
load_dotenv()
```

```python
data_path = os.environ.get("DATA_PATH")
df_train = pd.read_csv(data_path+"/train.csv", sep=';')
```

```python
df_train.columns
```

```python
facies_colors = ['#FFFF00', '#825000', '#FF7800',
                   '#00BEFF', '#00FFFF', '#783CA0',
                   '#006400', '#C878C8', '#FFDCFF',
                   '#000000', '#FF00FF', '#32EBB9']

sns.set_palette(sns.color_palette(facies_colors))
```

```python
sns_plot = sns.pairplot(df_train.drop(['WELL','DEPTH_MD','X_LOC','Y_LOC','Z_LOC','GROUP', 'FORMATION', 'FORCE_2020_LITHOFACIES_CONFIDENCE', 'CALI', 'RSHA', 'RMED', 'RDEP', 'SGR', 'BS', 'ROP', 'DCAL', 'DRHO', 'MUDWEIGHT', 'RMIC', 'ROPA', 'RXO'],axis=1),
             hue='FORCE_2020_LITHOFACIES_LITHOLOGY', palette=facies_colors)
plt.show()
#sns_plot.savefig('cross_plots.png')
```
